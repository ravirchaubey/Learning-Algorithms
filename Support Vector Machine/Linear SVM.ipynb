```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import scipy.io
df1 = scipy.io.loadmat('ex6data1.mat')

df1 = np.c_[df1['X'],df1['y']]
df1 = pd.DataFrame(df1,columns=['X1','X2','Y'])
df1.head()
```

```python
sns.set_style('whitegrid')
fig = plt.figure(dpi = 120,figsize=(3.5,2.5))
ax = fig.add_axes([1,1,1,1])

sns.scatterplot(x = 'X1',y = 'X2',hue = 'Y',data = df1,
               style = 'Y',palette=['yellow','black'])
handles,label = ax.get_legend_handles_labels()
ax.legend(handles,['Class 0','Class 1']);
```

```python
x = df1[['X1','X2']].values
y = df1['Y'].values
```

```python
print(x.shape)
print(y.shape)
```

## SVM without regularization

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC(kernel='linear')
svc_model.fit(x,y)
```

```python
print(svc_model.score(x,y))
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
sns.set_style('white')
fig = plt.figure(dpi = 120,figsize=(3.5,2.5))
ax = fig.add_axes([1,1,1,1])

plot_decision_regions(X=x,
                      y=y.astype('int'),
                      clf=svc_model,ax = ax)

ax.set(xlabel = 'X1',ylabel = 'X2')
ax.set_title('Linear Decision Boundry without Regularization',fontsize = 10)
handles,label = ax.get_legend_handles_labels()
ax.legend(handles,['Class 0','Class 1']);
```

## SVM with regularization

```python
svc_model = SVC(kernel='linear',C = 10)
svc_model.fit(x,y)
```

```python
sns.set_style('white')
fig = plt.figure(dpi = 120,figsize=(3.5,2.5))
ax = fig.add_axes([1,1,1,1])

plot_decision_regions(X=x,
                      y=y.astype('int'),
                      clf=svc_model,ax = ax)

ax.set(xlabel = 'X1',ylabel = 'X2')
ax.set_title('Linear Decision Boundry with Regularization (C=10)',fontsize = 9)
handles,label = ax.get_legend_handles_labels()
ax.legend(handles,['Class 0','Class 1']);
```

```python
svc_model = SVC(kernel='linear',C = 100)
svc_model.fit(x,y)
```
