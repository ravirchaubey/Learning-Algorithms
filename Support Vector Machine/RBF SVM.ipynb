```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import scipy.io
df1 = scipy.io.loadmat('ex6data2.mat')
df1 = np.c_[df1['X'],df1['y']]
df1 = pd.DataFrame(df1,columns=['X1','X2','Y'])
df1.head()
```

```python
sns.set_style('whitegrid')
fig = plt.figure(dpi = 120,figsize=(4,2.5))
ax = fig.add_axes([1,1,1,1])

sns.scatterplot(x = 'X1',y = 'X2',hue = 'Y',data = df1,
               style = 'Y',palette=['yellow','black'])
handles,label = ax.get_legend_handles_labels()
ax.legend(handles,['Class 0','Class 1']);
```

```python
x = df1[['X1','X2']].values
y = df1['Y'].values
```

```python
from sklearn.svm import SVC
svc_model = SVC(kernel='rbf')
svc_model.fit(x,y)
```

```python
from mlxtend.plotting import plot_decision_regions

fig = plt.figure(dpi = 120,figsize=(4,3))
ax = fig.add_axes([1,1,1,1])

plot_decision_regions(X=x,
                      y=y.astype('int'),
                      clf=svc_model,ax = ax,zoom_factor=1)
ax.set_xlim(0,1.2)
ax.set_ylim(0,1.2)
ax.set(xlabel = 'X1',ylabel = 'X2')
ax.set_title('Gausian Decision Boundry without Regularization',fontsize = 9)
handles,label = ax.get_legend_handles_labels()
ax.legend(handles,['Class 0','Class 1']);
```

```python
from sklearn.svm import SVC
svc_model = SVC(kernel='rbf',C = 10)
svc_model.fit(x,y)
```

```python
from mlxtend.plotting import plot_decision_regions

fig = plt.figure(dpi = 120,figsize=(4,3))
ax = fig.add_axes([1,1,1,1])

plot_decision_regions(X=x,
                      y=y.astype('int'),
                      clf=svc_model,ax = ax,zoom_factor=1)
ax.set_xlim(0,1.2)
ax.set_ylim(-0.2,1.2)
ax.set(xlabel = 'X1',ylabel = 'X2')
ax.set_title('Gausian Decision Boundry with Regularization (C=10)',fontsize = 9)
handles,label = ax.get_legend_handles_labels()
ax.legend(handles,['Class 0','Class 1']);
```

```python
from sklearn.svm import SVC
svc_model = SVC(kernel='rbf',C = 10,gamma=100)
svc_model.fit(x,y)
```

```python
from mlxtend.plotting import plot_decision_regions

fig = plt.figure(dpi = 120,figsize=(4,3))
ax = fig.add_axes([1,1,1,1])

plot_decision_regions(X=x,
                      y=y.astype('int'),
                      clf=svc_model,ax = ax,zoom_factor=1)
ax.set_xlim(0,1.2)
ax.set_ylim(0,1.2)
ax.set(xlabel = 'X1',ylabel = 'X2')
ax.set_title('Gausian Decision Boundry with Regularization (C=10, gamma = 100)',
             fontsize = 8)
handles,label = ax.get_legend_handles_labels()
ax.legend(handles,['Class 0','Class 1']);
```
